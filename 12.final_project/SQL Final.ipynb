## 3.2 Определение кол-ва обзоров и средней оценки каждой книги <a name="count2"></a>

Для каждой книги рассчитаем количество обзоров и среднюю оценку.

```python
#пишем запрос
query = '''
        select  b.book_id
                ,b.title
                ,count(rw.review_id) as count_reviews --расчет кол-ва обзоров на книгу
                ,avg(rt.rating) as avg_rate --расчет среднего рейтинга книги
        from books b
            left join ratings rt
                on b.book_id = rt.book_id
            left join reviews rw
                on b.book_id = rw.book_id
        group by b.book_id,b.title
'''
```

```python
#выполняем его
books_rate_reviews = pd.io.sql.read_sql(query, con = engine)
```

```python
#из любопытства выведем TOP-10 книг с наибольшим кол-вом обзоров
display(books_rate_reviews.sort_values(by = 'count_reviews', ascending = False).head(10))
```

```python
#из любопытства выведем TOP-10 книг с самой высокой средней оценкой среди книг, на которые написали более 50 обзоров
display(books_rate_reviews.query('count_reviews >= 50').sort_values(by = 'avg_rate', ascending = False).head(10))
```

## 3.3 Издательство с наибольшим кол-вом выпущенных больших книг <a name="count3"></a>

Определим издательство, которое выпустило наибольшее число книг толще 50 страниц, чтобы исключить из анализа брошюры.

```python
#пишем запрос
query = ''' 
                select 
                        p.publisher
                        ,count(b.book_id) as total_books
                from books b
                    left join publishers p
                        on b.publisher_id = p.publisher_id
                where b.num_pages > 50
                group by p.publisher
                order by total_books desc
'''
```

```python
#выполняем его
pd.io.sql.read_sql(query, con = engine).head(1)
```

## 3.4 Определение автора с самой высокой средней оценкой <a name="count4"></a>

Определим автора с самой высокой средней оценкой книг, учитывая только книги с 50 и более оценками.

```python
#пишем запрос
query = ''' 
        --берем книги с 50 и более оценками, считаем их средний рейтинг
        with t as (
        select  b.book_id
                ,count(rt.rating_id) as ratings
                ,avg(rt.rating) as avg_book_rate
        from books b
            left join ratings rt
                on b.book_id = rt.book_id
        group by b.book_id
        HAVING count(rt.rating_id) >= 50
        )
        
        --добавляем к предыдущей таблице авторов, рассчитываем среднюю оценку книг по автору
        select 
                a.author
                ,avg(t.avg_book_rate) as avg_author_rate  
        from t 
            left join books b
                on t.book_id = b.book_id
            left join authors a
                on b.author_id = a.author_id
        group by a.author
        order by avg_author_rate desc
'''
```

```python
#выполняем его
pd.io.sql.read_sql(query, con = engine).head(1)
```

**Вывод**

***J.K. Rowling/Mary GrandPré*** - автор с самой высокой средней оценкой книг (равной **4.28**) среди книг с 50 и более оценками.

## 3.5 Расчет среднего кол-ва обзоров по пользователям <a name="count5"></a>

Посчитаем среднее количество обзоров от пользователей, которые поставили больше 50 оценок.

```python
#пишем запрос
query = '''
        select round(AVG(sub.total_reviews)) as avg_reviews
        from 
        (select  rw.username
                ,count(rw.review_id) total_reviews
        from reviews rw
        where rw.username in (
            --отбираем пользователей, которые поставили более 50 оценок
            select rt.username
            from ratings rt
            group by rt.username
            having count(rt.rating_id) > 50
            )
        group by rw.username) as sub
'''
```

```python
#выполняем его
pd.io.sql.read_sql(query, con = engine)
```
