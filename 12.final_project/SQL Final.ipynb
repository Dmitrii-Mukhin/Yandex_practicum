–î–ª—è –∫–∞–∂–¥–æ–π –∫–Ω–∏–≥–∏ —Ä–∞—Å—Å—á–∏—Ç–∞–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –æ–±–∑–æ—Ä–æ–≤ –∏ —Å—Ä–µ–¥–Ω—é—é –æ—Ü–µ–Ω–∫—É.

```python
#–ø–∏—à–µ–º –∑–∞–ø—Ä–æ—Å
query = '''
        select  b.book_id
                ,b.title
                ,count(rw.review_id) as count_reviews --—Ä–∞—Å—á–µ—Ç –∫–æ–ª-–≤–∞ –æ–±–∑–æ—Ä–æ–≤ –Ω–∞ –∫–Ω–∏–≥—É
                ,avg(rt.rating) as avg_rate --—Ä–∞—Å—á–µ—Ç —Å—Ä–µ–¥–Ω–µ–≥–æ —Ä–µ–π—Ç–∏–Ω–≥–∞ –∫–Ω–∏–≥–∏
        from books b
            left join ratings rt
                on b.book_id = rt.book_id
            left join reviews rw
                on b.book_id = rw.book_id
        group by b.book_id,b.title
'''
```

```python
#–≤—ã–ø–æ–ª–Ω—è–µ–º –µ–≥–æ
books_rate_reviews = pd.io.sql.read_sql(query, con = engine)
```

```python
#–∏–∑ –ª—é–±–æ–ø—ã—Ç—Å—Ç–≤–∞ –≤—ã–≤–µ–¥–µ–º TOP-10 –∫–Ω–∏–≥ —Å –Ω–∞–∏–±–æ–ª—å—à–∏–º –∫–æ–ª-–≤–æ–º –æ–±–∑–æ—Ä–æ–≤
display(books_rate_reviews.sort_values(by = 'count_reviews', ascending = False).head(10))
```

## 3.4 –û–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –∞–≤—Ç–æ—Ä–∞ —Å —Å–∞–º–æ–π –≤—ã—Å–æ–∫–æ–π —Å—Ä–µ–¥–Ω–µ–π –æ—Ü–µ–Ω–∫–æ–π <a name="count4"></a>

–û–ø—Ä–µ–¥–µ–ª–∏–º –∞–≤—Ç–æ—Ä–∞ —Å —Å–∞–º–æ–π –≤—ã—Å–æ–∫–æ–π —Å—Ä–µ–¥–Ω–µ–π –æ—Ü–µ–Ω–∫–æ–π –∫–Ω–∏–≥, —É—á–∏—Ç—ã–≤–∞—è —Ç–æ–ª—å–∫–æ –∫–Ω–∏–≥–∏ —Å 50 –∏ –±–æ–ª–µ–µ –æ—Ü–µ–Ω–∫–∞–º–∏.

```python
#–ø–∏—à–µ–º –∑–∞–ø—Ä–æ—Å
query = ''' 
        --–±–µ—Ä–µ–º –∫–Ω–∏–≥–∏ —Å 50 –∏ –±–æ–ª–µ–µ –æ—Ü–µ–Ω–∫–∞–º–∏, —Å—á–∏—Ç–∞–µ–º –∏—Ö —Å—Ä–µ–¥–Ω–∏–π —Ä–µ–π—Ç–∏–Ω–≥
        with t as (
        select  b.book_id
                ,count(rt.rating_id) as ratings
                ,avg(rt.rating) as avg_book_rate
        from books b
            left join ratings rt
                on b.book_id = rt.book_id
        group by b.book_id
        HAVING count(rt.rating_id) >= 50
        )
        
        --–¥–æ–±–∞–≤–ª—è–µ–º –∫ –ø—Ä–µ–¥—ã–¥—É—â–µ–π —Ç–∞–±–ª–∏—Ü–µ –∞–≤—Ç–æ—Ä–æ–≤, —Ä–∞—Å—Å—á–∏—Ç—ã–≤–∞–µ–º —Å—Ä–µ–¥–Ω—é—é –æ—Ü–µ–Ω–∫—É –∫–Ω–∏–≥ –ø–æ –∞–≤—Ç–æ—Ä—É
        select 
                a.author
                ,avg(t.avg_book_rate) as avg_author_rate  
        from t 
            left join books b
                on t.book_id = b.book_id
            left join authors a
                on b.author_id = a.author_id
        group by a.author
        order by avg_author_rate desc
'''
```

```python
#–≤—ã–ø–æ–ª–Ω—è–µ–º –µ–≥–æ
pd.io.sql.read_sql(query, con = engine).head(1)
```

## 3.5 –†–∞—Å—á–µ—Ç —Å—Ä–µ–¥–Ω–µ–≥–æ –∫–æ–ª-–≤–∞ –æ–±–∑–æ—Ä–æ–≤ –ø–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è–º <a name="count5"></a>

–ü–æ—Å—á–∏—Ç–∞–µ–º —Å—Ä–µ–¥–Ω–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –æ–±–∑–æ—Ä–æ–≤ –æ—Ç –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π, –∫–æ—Ç–æ—Ä—ã–µ –ø–æ—Å—Ç–∞–≤–∏–ª–∏ –±–æ–ª—å—à–µ 50 –æ—Ü–µ–Ω–æ–∫.

```python
#–ø–∏—à–µ–º –∑–∞–ø—Ä–æ—Å
query = '''
        select round(AVG(sub.total_reviews)) as avg_reviews
        from 
        (select  rw.username
                ,count(rw.review_id) total_reviews
        from reviews rw
        where rw.username in (
            --–æ—Ç–±–∏—Ä–∞–µ–º –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π, –∫–æ—Ç–æ—Ä—ã–µ –ø–æ—Å—Ç–∞–≤–∏–ª–∏ –±–æ–ª–µ–µ 50 –æ—Ü–µ–Ω–æ–∫
            select rt.username
            from ratings rt
            group by rt.username
            having count(rt.rating_id) > 50
            )
        group by rw.username) as sub
'''
```

```python
#–≤—ã–ø–æ–ª–Ω—è–µ–º –µ–≥–æ
pd.io.sql.read_sql(query, con = engine)
```
